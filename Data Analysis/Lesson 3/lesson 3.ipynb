```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.DataFrame()
first = pd.read_excel('./first.xlsx')
second = pd.read_excel('./second.xlsx')
all_data = all_data.append(first,ignore_index=True)
all_data = pd.merge(all_data, second, how='outer').drop_duplicates()
#random_data = all_data.sample(frac=1)
```

```python
def get_column(data, column_name):
    new_data = data[column_name]
    new_random_data = new_data.sample(frac=1)
    
    return new_data, new_random_data
```

```python
def get_value_counts(column):
    return column.value_counts().to_dict()
```

```python
def fill_data(all_data, columns):
    data = {}
    data_shuffled = {}
    for column in columns:
        column_curr, column_curr_shuffled = get_column(all_data, column)

        data[column] = get_value_counts(column_curr)
        data_shuffled[column] = get_value_counts(column_curr_shuffled)
    
    return data, data_shuffled
```

```python
columns = all_data.columns

data = []
for i in range(0, len(all_data), 100):
    batch = all_data.iloc[i:i+100]
    data_curr, data_shuffled = fill_data(batch, columns)
    data.append(data_curr)
```

```python
col_1 = all_data['Подразделение 4']
col_2 = all_data['Подразделение 5']
col_3 = all_data['Подразделение 6']
col_4 = all_data['Подразделение 7']

col_1_categories = col_1.unique()
col_2_categories = col_2.unique()
col_3_categories = col_3.unique()
col_4_categories = col_4.unique()

value_pairs_1_2 = list(zip(col_1, col_2))
value_pairs_1_3 = list(zip(col_1, col_3))
value_pairs_1_4 = list(zip(col_1, col_4))
value_pairs_2_3 = list(zip(col_2, col_3))
value_pairs_2_4 = list(zip(col_2, col_4))
value_pairs_3_4 = list(zip(col_3, col_4))

value_pairs = [value_pairs_1_2,
               value_pairs_1_3,
               value_pairs_1_4,
               value_pairs_2_3,
               value_pairs_2_4,
               value_pairs_3_4,]
```

```python
from collections import Counter

matrix_data = Counter()

for pair in value_pairs:
    for value in pair:
        t1 = str(value[0])
        t2 = str(value[1])
        matrix_data[(t1, t2)] += 1
```

```python
categories = np.concatenate((col_1_categories, col_2_categories, col_3_categories, col_4_categories))
print(categories[1:])
categories = np.unique(categories[1:])
matrix_values_count = pd.DataFrame(0,index=categories, columns=categories)

for i in matrix_data:
    index = i[1]
    column = i[0]
    matrix_values_count.loc[index, column] = matrix_data[i]

matrix_values_count = matrix_values_count.fillna(0).apply(np.int64)
```
